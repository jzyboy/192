```python
from selenium import webdriver
driver=webdriver.Chrome()
url = 'https://map.baidu.com/search/%E9%85%92%E5%BA%97/@13401816.6,3523321.62,17z?querytype=nb&r=1000&wd=%E9%85%92%E5%BA%97&nb_x=13401816.60&nb_y=3523321.62&uid=7a0cb8c0d67a634683a00c28&b=(13400809.909611505,3522782.19066226;13402823.09336229,3524093.869264365)&l=17&gr_radius=1000&pn=0&device_ratio=2&da_src=shareurl'
driver.get(url)
html=driver.page_source
```

```python
import re
import pandas as pd
list1=[]
list2=[]
list3=[]
html = driver.page_source
for i in range(2):
    html = driver.page_source
    reobj = re.compile(r'<a href="javascript:void\(0\)"[\d\D]*?data-index=[\d\D]*?data-detail-code="list_page">([\d\D]*?)</a>[\d\D]*?<span>                    [\d\D]*?<i class="seperate">\|</i>([\d\D]*?)</span>[\d\D]*?<div class="row addr">                <span class="n-grey" title="([\d\D]*?)">', re.MULTILINE)
    for match in reobj.finditer(html):
	# match start: match.start()
	# match end (exclusive): match.end()
	# matched text: match.group()
        xingming=match.group(1)
        leixin=match.group(2)
        dizhi=match.group(3)
        list1.append(xingming)
        list2.append(leixin)
        list3.append(dizhi)
    driver.find_element_by_link_text(u"下一页>").click()
    import time
    time.sleep(5)
pd.DataFrame({'店名':list1,'类型':list2,'地址':list3})
```

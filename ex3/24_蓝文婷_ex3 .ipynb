```python
from selenium import webdriver
driver=webdriver.Chrome()
url = 'https://www.baidu.com/s?wd=%E6%98%8E%E6%98%9F&ie=utf-8&tn=78040160_14_pg&ch=16'
driver.get(url)
```

```python
html = driver.page_source
#print(html)
```

```python
import re
import time
list=[]
html = driver.page_source
for i in range(330):
    html = driver.page_source
    reobj = re.compile(r'<p class="op_exactqa_item_img"><a href=".*?" title=".*?" target="_blank"><img class="c-img c-img4" src="([\s\S]*?)"></a></p><p class="c-gap-top-small"><a href="([\s\S]*?)" title=".*?" target="_blank">([\s\S]*?)</a></p>')
    for match in reobj.finditer(html):
        a=match.group(1)
        b=match.group(2)
        c=match.group(3)
        d=(c,b,a)
        list.append(d)
    driver.find_element_by_xpath("//div[@id='1']/div/div/div/div[2]/p/span[4]").click()
    driver.find_element_by_xpath("//div[@id='1']/div/div/div/div[2]/p[2]/span[3]").click()
    driver.find_element_by_xpath("//div[@id='1']/div/div/div[2]/div[2]/p/span[6]").click()
    time.sleep(2)
print(list)
```

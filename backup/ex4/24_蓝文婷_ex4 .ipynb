```python
from selenium import webdriver
driver = webdriver.Chrome()
url = "https://www.yinxiang.com/everhub/"
driver.get(url)
```

```python
html = driver.page_source#获取源代码
#print (html)
```

```python
import re
import time
list=[]
html = driver.page_source
for i  in range(50):
    html = driver.page_source
    reobj = re.compile(r'<span class="NoteCard_userName__15MR_">([\s\S]*?)</span></div><div class="NoteCard_content__s8T-9 NoteCard_notIE__1YX8b"><div class="NoteCard_noteInfo__3mBWD"><div><div><h3 class="NoteCard_title__juR5k">([\s\S]*?)</h3></div><div><p class="NoteCard_description__1jOUe NoteCard_twoRows__22BR2">([\s\S]*?)</p></div></div>')
    for match in reobj.finditer(html):
        a=match.group(1)
        b=match.group(2)
        c=match.group(3)
        t=(a,c)
        list.append(t)
    from selenium.webdriver.common.keys import Keys    #键盘导入类
    driver.find_element_by_xpath('//html').send_keys(Keys.SPACE)
    time.sleep(2)
print (list)
```
